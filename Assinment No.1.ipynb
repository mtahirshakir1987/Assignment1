### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

```python
# Write your code here
weight_kgs = float(input("Please input your weight in KGS : "))
hight_inches = float(input("Please input your height in Inches :"))

hight_meter = hight_inches*0.0254

bmi = int(weight_kgs / hight_meter**2)

"""
BMI Categories
Underweight (kam wazan) → BMI 18.5 se kam
Normal weight (normal) → BMI 18.5 – 24.9
Overweight (zayada wazan) → BMI 25 – 29.9
Obese (motapa) → BMI 30 ya us se zyada
"""
if bmi>=30:
    bmi_category = "Obese (motapa)"
elif bmi>=25:
    bmi_category = "Overweight (zayada wazan)"
elif bmi>=18.5:
    bmi_category = "Normal Weight (normal)"
else:
    bmi_category = "Underweight (kam wazan)"

print (f"Your BMI is {bmi}, and as per BMI Category you are {bmi_category}")
```
